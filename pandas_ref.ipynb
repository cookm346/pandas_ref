Filtering

```
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```
df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
```

```
#reset index for filtered data
df = df.reset_index()
df = df.reset_index(drop=True) #don't make new col with old indices
df.reset_index(drop=True, inplace=True) #don't need to reassign to new object
```

```
df.loc[df["Name"].str.contains("Mega")]
```

```
df = df.loc[~df["Name"].str.contains("Mega")] #tilde for not
```

```
df.loc[df["Type 1"].str.contains("Fire|Grass", regex=True)]
```

```
#import regex library and ignore case
import re
df.loc[df["Type 1"].str.contains("fire|grass", flags=re.I, regex=True)]
```

```
df.loc[df["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)] #ind all names that start w/ Pi
```

```
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flame"
```

Aggregate Stats

```
df = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv")
```

```
df.groupby(["Type 1"]).mean()
```

```
df.groupby(["Type 1"]).mean().sort_values("Defense", ascending=False)
```

```
df.groupby(["Type 1", "Type 2"]).mean()
```
